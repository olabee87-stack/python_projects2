```python
# Countries data

countries_data = pd.read_csv('countries.csv')
print(countries_data[:51])
```

```python
# Simple analysis - First step, extract the contries you wish to analyse

afghan = countries_data[countries_data.country == "Afghanistan"]
albania = countries_data[countries_data.country == "Albania"]

# Print both 
print(f"{afghan}" "\n" f"{albania}")
```

```python
# 2nd step, Analyse the population and year of Afghan and Albania 
plt.plot(afghan.year, afghan.population /10**6) #the division is to show the pop in millions
plt.plot(albania.year, albania.population / 10 ** 6)

plt.title('Population by the year')

plt.xlabel("X-AXIS (YEAR)")
plt.ylabel("Y-AXIS (POPULATION)")

plt.legend(["Afghan", "Albania"])
plt.show()
```

```python
# Show the percentage growth from the first year
# Step 1 - Make the first year 100% and then divide the rest of the population by the first one

afghan.population /afghan.population[0] * 100
```

```python
# Check both countries' percent against each other (iloc - to avoid the print sometimes)
plt.plot(afghan.year, afghan.population / afghan.population.iloc[0] * 100) #the division is to show the pop in millions
plt.plot(albania.year, albania.population / albania.population.iloc[0] * 100)

plt.title('Population by the year')

plt.xlabel("X-AXIS (YEAR)")
plt.ylabel("Y-AXIS (POPULATION GROWTH) - FIRST YEAR = 100")

plt.legend(["Afghan", "Albania"])
plt.show()
```
